```python
from sklearn import tree
import pandas as pd
import dataframe_image as dfi
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(os.path.join("data", "doge.csv"))
crypto = pd.read_csv(os.path.join("data", "price.csv"))
df = df.rename(columns={"Datetime":"date"})
```

```python
sentiment0=[]
for i in range(len(df)):
    if df.loc[i,'negative']>0:
        sentiment0+=[0]
    elif df.loc[i,'positive']>0:
        sentiment0+=[2]        
    else:
        sentiment0+=[1]    
```

```python
df['sentiment']=sentiment0
df['sentiment'].value_counts()

df.head()
```

```python
crypto.reset_index(inplace=True)
crypto = crypto.drop(['index'], axis = 1)
crypto['date'] = pd.to_datetime(crypto['date'])
crypto['date'] = crypto['date'].dt.strftime('%Y-%m-%d')
crypto
```

```python
eth_sentiment = df.groupby('date')['sentiment'].mean()
eth_sentiment = eth_sentiment.reset_index()
eth_sentiment
```

```python
eth_price_sen = pd.merge(eth_sentiment, crypto,how='outer', on='date')
eth2 = eth_price_sen.drop(columns=["Unnamed: 0"], axis = 1)
```

```python
eth_clean = eth2.dropna(subset=['sentiment'])
```

```python
eth_clean
```

```python
#Change the price reference to the relevant Crypto you want to analyse
eth_clean['Change DOGE'] = eth_clean.dogecoin_price - eth_clean.dogecoin_price.shift(1)
eth_clean
```

```python
eth_clean['DOGE Movement'] = np.where(eth_clean['Change DOGE']>0, 1, 0)
eth_clean
```

```python
#update the target to the right crypto
target = eth_clean["DOGE Movement"]
target_names = ["sentiment", "dogecoin_price"]
```

```python
data = eth_clean.drop(["DOGE Movement", "date", "Change DOGE"], axis=1)
feature_names = data.columns
data.head()
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# Create a Decision Tree Classifier
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# Fit the classifier to the data
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# Bonus
# Create, fit, and score a Random Forest Classifier
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
final = pd.DataFrame(sorted(zip(rf.feature_importances_, feature_names), reverse=True))
final
```

```python
final.columns = ["Weighting %", "Variable"]
final
```
