```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join("..","..","..", "data", "ethereum_cleaned.csv"))
crypto = pd.read_csv(os.path.join("..","..","..","data", "price.csv"))
df = df.rename(columns={"Datetime":"date"})
```

```python
sentiment0=[]
for i in range(len(df)):
    if df.loc[i,'negative']>0:
        sentiment0+=[0]
    elif df.loc[i,'positive']>0:
        sentiment0+=[2]        
    else:
        sentiment0+=[1]    
```

```python
df['sentiment']=sentiment0
df['sentiment'].value_counts()

df.head()
```

```python
crypto.reset_index(inplace=True)
crypto = crypto.drop(['index'], axis = 1)
crypto['date'] = pd.to_datetime(crypto['date'])
crypto['date'] = crypto['date'].dt.strftime('%Y-%m-%d')
crypto
```

```python
eth = pd.merge(df, crypto,how='outer', on='date')
```

```python
eth
```

```python
target = df["ETH Up/Down"]
target_names = ["comp", "ethereum_price"]
```

```python
data = df.drop(["ETH Up/Down", "date", "Unnamed: 0", "Change ETH", "Pos/Neg"], axis=1)
feature_names = data.columns
data.head()
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# Create a Decision Tree Classifier
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# Fit the classifier to the data
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# Bonus
# Create, fit, and score a Random Forest Classifier
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
